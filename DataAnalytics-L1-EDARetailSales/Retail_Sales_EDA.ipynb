```python
category_sales = df.groupby("Product Category")["Total Amount"].sum()

category_sales
```

```python
plt.figure(figsize=(8,5))

category_sales.plot(kind="bar")

plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")

plt.show()
```

### Insight

- Electronics has the highest total sales.
- Clothing is the second-highest category.
- Beauty has the lowest sales.

```python
gender_sales = df.groupby("Gender")["Total Amount"].sum()

gender_sales
```

```python
plt.figure(figsize=(6,4))

gender_sales.plot(kind="bar")

plt.title("Total Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Sales")

plt.show()
```

### Insight

- Female customers generated higher total sales than male customers.
- This suggests that female customers contribute more to overall revenue.
- The company can consider targeted marketing campaigns for female customers while exploring ways to increase engagement among male customers.

```python
plt.figure(figsize=(8,5))

plt.hist(df["Age"], bins=10)

plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Number of Customers")

plt.show()
```

### Insight

- Most customers belong to the 25–40 age group.
- Younger adults contribute a large portion of the customer base.
- The company can target this age group with personalized promotions and loyalty programs.

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Month"] = df["Date"].dt.month_name()
```

```python
monthly_sales = df.groupby("Month")["Total Amount"].sum()

monthly_sales
```

```python
plt.figure(figsize=(10,5))

monthly_sales.plot(kind="line", marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)
plt.show()
```

### Insight

- Sales increased during the middle of the year.
- Some months recorded lower sales than others.
- The business can analyze the reasons for peak and low sales months to improve future planning.

```python
top_transactions = df.nlargest(10, "Total Amount")

top_transactions
```

### Insight

- The table shows the top 10 highest-value transactions.
- These customers made the largest purchases.
- The business can reward such customers through loyalty programs or exclusive offers.

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Blues")

plt.title("Correlation Heatmap")

plt.show()
```

### Insight

- The heatmap shows the relationship between numerical variables.
- Quantity and Total Amount often have a positive correlation because purchasing more items generally increases the total bill.
- Correlation helps identify which variables influence each other.

# Final Business Insights
